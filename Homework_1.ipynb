```python
# Task 1


# Historical Google stock price

import pandas_datareader as pdr


api_key='71be671ca49c87ed644956fdd46f8a398b51f899' 

start="2020-1-1"
end="2021-1-1"

df = pdr.tiingo.TiingoDailyReader('GOOG', start=start, end=end, api_key=api_key)


goog = df.read()

goog
```

```python
# Saving results to CSV file


import pandas as pd

goog.to_csv('goog_hist_price.csv', date_format='%Y-%m-%d')


# Reading the results

google_df = pd.read_csv('goog_hist_price.csv', header=0, index_col='date', parse_dates=True)

google_df
```

```python
# Plotting results

import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline

google_df['close'].plot(grid=True,)
```
